```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import math
```

```python
#change the directory to read csv file
os.chdir(r"C:\Users\tao_1\OneDrive\Documents\Fall2021\ABE305_Final_Project\Data")

#read sample data
df = pd.read_csv("272.csv", skiprows=7)
df = df.rename(columns={'mm': 'Extension', 'N': 'Load'})

plt.plot(df['Extension'], df['Load'], "ro", markersize= 3)
plt.title("Load vs. Extension")
plt.xlabel("Extension (mm)")
plt.ylabel("Load")
plt.grid(5)
plt.show()
```

```python
list_slope_1 = []
list_ext_1 = []

for i in range(len(df)-1):
    slope = (df['Load'][i+1] - df['Load'][i]) / (df['Extension'][i+1] - df['Extension'][i])
    list_slope_1.append(slope)
    list_ext_1.append(df['Extension'][i+1])
    
    #sort to data frame 
df_slope_1 = {'Slope':list_slope_1,'Extension':list_ext_1}
df_slope_1 = pd.DataFrame(df_slope_1)

#first deravative 

#polyfit for firs deravative
coefficients_1 = np.polyfit(df_slope_1['Extension'], df_slope_1['Slope'], 5)

#Calculate the polynomial
poly_1 = np.poly1d(coefficients_1)

x = df_slope_1['Extension']
y = df_slope_1['Slope']
new_x = np.arange(0,30,0.1)
new_y = poly_1(new_x)


plt.plot(x, y, "o", new_x, new_y, markersize= 3)
plt.title("First slope vs. Extension")
plt.xlabel("Extension (mm)")
plt.ylabel("first slope")
plt.ylim([-15,15])
plt.grid(5)
plt.show()
```

```python
list_slope_2 = []
list_ext_2 = []

for i in range(len(df_slope_1)-1):
    slope = (df_slope_1['Slope'][i+1] - df_slope_1['Slope'][i]) / (df_slope_1['Extension'][i+1] - df_slope_1['Extension'][i])
    list_slope_2.append(slope)
    list_ext_2.append(df_slope_1['Extension'][i+1])
    
    #sort to data frame 
df_slope_2 = {'Slope':list_slope_2,'Extension':list_ext_2}
df_slope_2 = pd.DataFrame(df_slope_2)

#second deravative 

#polyfit for firs deravative
coefficients_2 = np.polyfit(df_slope_2['Extension'], df_slope_2['Slope'], 5)

#Calculate the polynomial
poly_2 = np.poly1d(coefficients_2)

x = df_slope_2['Extension']
y = df_slope_2['Slope']
new_x = np.arange(0,30,0.1)
new_y = poly_2(new_x)


plt.plot(x, y, "o", new_x, new_y, markersize= 3)
plt.title("second slope vs. Extension")
plt.xlabel("Extension (mm)")
plt.ylabel("second slope")
plt.grid(5)
plt.ylim([-15,15])
plt.show()
```
